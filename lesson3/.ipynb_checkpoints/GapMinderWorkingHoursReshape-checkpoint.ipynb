```python
import pandas as pd
import numpy as np
```

```python
#read in the dataframe
df = pd.read_excel('indicator_hours per week.xlsx')
df.head()
```

```python
#rename the first column
columns = list(df.columns)
columns[0] = 'country'
df.columns = columns
df.head()
```

```python
#setting index to country
dfreindex = df.set_index(['country'])
dfreindex.head()
```

```python
#transposing the data frame to do a fillna by mean of country (because doing row by row is complex and ineffcient)
transposed = dfreindex.transpose()
transposed.head()
```

```python
#now fill with the mean
filled = transposed.fillna(transposed.mean())
filled.head()
```

```python
#transpose it back
clean = filled.transpose()
clean.head()
```

```python
#now reset the index back to the old index
clean = clean.reset_index()
clean.head()
```

```python
#switch from wide to long using melt
melt = pd.melt(clean, id_vars= ['country'], var_name= 'year')
melt.head()
```
